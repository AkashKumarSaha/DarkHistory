```python
%autoreload

import matplotlib
matplotlib.rc_file('matplotlibrc')
import matplotlib.pyplot as plt

%matplotlib inline

matplotlib.rcParams['figure.figsize'] = [10,10]


import numpy as np
import darkhistory.spec.pppc as pppc
```

```python
eng = 10**np.arange(-4, 12.5, 16.5/500)

mDM = 49e9
chan = 'b'

test_phot = pppc.get_pppc_spec(mDM, eng, chan, 'phot', decay=False)
test_elec = pppc.get_pppc_spec(mDM, eng, chan, 'elec', decay=False)

print('Fraction of energy in photons: ', test_phot.toteng()/(2*mDM )) 
print('Fraction of energy in e+e-: ', test_elec.toteng()/(2*mDM))
print('Fraction of energy in EM states: ', test_phot.toteng()/(2*mDM) + test_elec.toteng()/(2*mDM))
```

```python

plt.figure()

plt.loglog()

phot_plot, = plt.plot(eng, test_phot.dNdE, label='photons')
elec_plot, = plt.plot(eng, test_elec.dNdE, label=r'$e^+e^-$')

plt.xlabel('Energy [eV]')
plt.ylabel(r'dN/dE [eV$^{-1}$]')
plt.title('Secondary Spectra')
plt.legend(handles=[phot_plot, elec_plot],loc=2)

plt.axis([1e-1, 4e12, 1e-15, 1e-4])
```
